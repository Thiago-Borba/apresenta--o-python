## Importando os dados para o python

A Library **pandas** tem funções para que se possa importar arquivos. Esses arquvivos são importados para um objeto chamado DataFrame, que é algo parecido com o que vemos no excel, utilizando um index para as linhas e nomes para as colunas.

A função para importar arquivos csv é: **pd.read_csv('destino do arquivo')**

Esse função possui uma serie argumentos pré definidos, um deles é **sep**, que define qual caracter será usado para separar as colunas do arquivo csv, como default ele é definido por ",", ou seja, o sistema americano. Como o arquivo exemplo não está no padrão americano, vamos definir o sep = ";"

O método **.head()** mostra as 5 primeiras linhas do data frame criado.



```python
## importando o arquivo csv
journal = pd.read_csv("https://raw.githubusercontent.com/Thiago-Borba/apresenta--o-python/master/Journal.csv", sep = ";")
```

```python
# mostrando o data frame
##  Para vizualizar todas as colunas use: pd.set_option("display.max_rows", None, "display.max_columns", None)
journal.head()
```

## Manipulando os dados

Também é possivel mudar "mexer" nos dados, mudando os números, calculando novas colunas e agrupando valores.

Para dar exemplo será calculada a coluna "leitura aprazivel", que é a média de P5a(Lê-se facilmente), p5b(leitura agradável) e p5c(Tem boa ilustração)

Para criar a coluna "leitura aprazivel" é necessário primeiro transformar os valores ordnais em continuos, para isso serão feitos os seguintes passos:
 1. criar um data frame com as colunas a serem alterados
 2. substitiur valores ordinais por continuos
 3. criar uma coluna que é a média desses valores

```python
# criar um data frame com as colunas a serem alterados

vari = ["p5a", "p5b", "p5c"]
journal_b = journal[vari]
journal_b.head()
```

```python
# substitiur valores ordinais por continuos

for i in vari:
  dic = {i:
         {"Discordo completamente":1,
          "Discordo":2,
          "Nao concordo nem discordo":3,
          "Concordo":4,
          "Concordo completamente":5}}
  journal_b.replace(dic, inplace = True)
```

```python
journal_b.head()
```

```python
#criar uma coluna que é a média desses valores

journal["leitura_aprezivel"] = (journal_b["p5a"] +  journal_b["p5b"] + journal_b["p5c"])/3
journal.head()
```

## Explorando os dados

A biblioteca **pandas** possui alguns comandos para facilitar a exploração dos dados, exemplo do **.describe** e **crosstab**
Também é possível usar o **.groupby**.

O comando **.describe** mostra as propriedades estatística de todos as colunas que possuem valores int ou float
O comando **.crosstab** cruze os dados de duas colunas diferentes
O comando **groupby** cria um novo dataframe agrupado por uma coluna.

```python
journal.describe()
```

```python
# usando o methdo crosstab para criar uma tabela cruzada
crosstab_1 = pd.crosstab(journal["Sexo"], journal["p1"], margins=True, normalize = "columns")
crosstab_1
```

```python
#usando o methodo grouby para agrupar os valores de journal e mostrar a média de leitura aprezivel de cada um deles
journal.groupby('p1').agg({'leitura_aprezivel':'mean'})
```

## Plotando gráficos

Com o auxílio da biblioteca **matplotlib** é possivel criar diversos gráficos. nessa demonstração será criado um gráfito tipo boxplot usando o método **.boxplot()**

```python
# o método boxplot possui o argumento by que utiliza uma das colunas selecionadas para separar os gráfios
# o argumento figsize altero o tamanho do gráfico
journal[["leitura_aprezivel", "p1"]].boxplot(by = "p1", figsize=(12, 8))

```
